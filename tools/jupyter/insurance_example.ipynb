> Задача: обработать базу клиентов, рассчитать премии, найти группу риска.

## Шаг 1: Данные клиентов

```python
# Список клиентов — список словарей
clients = [
    {'id': 1, 'name': 'Иванов А.П.',  'age': 35, 'accidents': 0, 'city': 'Москва'},
    {'id': 2, 'name': 'Петрова М.С.', 'age': 22, 'accidents': 1, 'city': 'СПб'},
    {'id': 3, 'name': 'Сидоров К.Д.', 'age': 47, 'accidents': 0, 'city': 'Москва'},
    {'id': 4, 'name': 'Орлова Е.В.',  'age': 19, 'accidents': 0, 'city': 'Казань'},
    {'id': 5, 'name': 'Козлов В.И.',  'age': 60, 'accidents': 2, 'city': 'Москва'},
]
print(f'Клиентов в базе: {len(clients)}')
print(f'Первый клиент: {clients[0]}')
```

## Шаг 2: Функция расчёта премии

```python
BASE_RATE = 12_000

def calculate_premium(age: int, accidents: int = 0) -> float:
    premium = BASE_RATE
    
    # Возрастной коэффициент
    if age < 25:
        premium *= 1.5   # молодые — риск выше
    elif age > 55:
        premium *= 1.3   # пожилые
    
    # Коэффициент аварийности
    if accidents > 0:
        premium *= 1 + accidents * 0.2
    
    return round(premium, 2)

# Тест
print('Молодой без аварий:', calculate_premium(22))        # 18000.0
print('Молодой 1 авария: ', calculate_premium(22, 1))     # 21600.0
print('Взрослый:         ', calculate_premium(35))        # 12000.0
print('Пожилой 2 аварии:', calculate_premium(60, 2))     # 20280.0
```

## Шаг 3: Применяем к базе (list comprehension)

```python
# Добавляем поле premium каждому клиенту
clients_with_premium = [
    {**c, 'premium': calculate_premium(c['age'], c['accidents'])}
    for c in clients
]

# Выводим красиво
print(f"{'Имя':<20} {'Возраст':>8} {'Аварии':>8} {'Премия':>12}")
print('-' * 52)
for c in clients_with_premium:
    print(f"{c['name']:<20} {c['age']:>8} {c['accidents']:>8} {c['premium']:>12,.0f}")
```

## Шаг 4: Статистика и группа риска

```python
premiums = [c['premium'] for c in clients_with_premium]

print('Статистика:')
print(f'  Клиентов:       {len(premiums)}')
print(f'  Мин. премия:    {min(premiums):,.0f} руб.')
print(f'  Макс. премия:   {max(premiums):,.0f} руб.')
print(f'  Средняя:        {sum(premiums)/len(premiums):,.0f} руб.')
print(f'  Итого сборов:   {sum(premiums):,.0f} руб.')

# Группа риска — клиенты с премией выше средней
avg = sum(premiums) / len(premiums)
high_risk = [c for c in clients_with_premium if c['premium'] > avg]

print(f'\nГруппа риска ({len(high_risk)} чел.):')
for c in high_risk:
    print(f"  {c['name']}: {c['premium']:,.0f} руб.")
```

## Шаг 5: Сохраняем в JSON

```python
import json

# Сохраняем
with open('clients_premiums.json', 'w', encoding='utf-8') as f:
    json.dump(clients_with_premium, f, ensure_ascii=False, indent=2)

print('Сохранено в clients_premiums.json')

# Читаем обратно для проверки
with open('clients_premiums.json', 'r', encoding='utf-8') as f:
    loaded = json.load(f)

print(f'Прочитано: {len(loaded)} клиентов')
print(json.dumps(loaded[0], ensure_ascii=False, indent=2))
```
